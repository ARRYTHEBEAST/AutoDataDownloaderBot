```python
import pygame
import pyautogui
import time
import sys
from tqdm import tqdm

# Initialize Pygame
pygame.init()

# Set the window size and position
win = pygame.display.set_mode((300, 100))  # Increased width for progress bar
pygame.display.set_caption("Progress Tracker")

# Define colors
WHITE = (255, 255, 255)
BLACK = (0, 0, 0)
GREEN = (0, 255, 0)

# Set up font
font = pygame.font.SysFont("Arial", 18)

# Function to update display with a progress bar
def update_display(year, month, state, commodity, progress):
    win.fill(WHITE)
    text_surface = font.render(f"Y:{year} M:{month} S:{state} C:{commodity}", True, BLACK)
    win.blit(text_surface, (10, 10))

    # Draw the progress bar
    pygame.draw.rect(win, BLACK, (10, 40, 280, 30), 2)  # Outline of the progress bar
    pygame.draw.rect(win, GREEN, (10, 40, 280 * progress, 30))  # Fill based on progress

    pygame.display.flip()

# Delay to prepare for the script to start
time.sleep(30)

# Calculate total number of iterations for progress tracking
total_iterations = 24 * 12 * 26 * 459
current_iteration = 0

# Main loop to handle events
running = True
while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

    # Nested loops with tqdm progress tracking
    for year in tqdm(range(1, 25), desc="Year Progress"):
        for month in range(1, 13):
            for state in range(1, 27):
                for commodity in range(1, 460):
                    # Update Pygame display with current loop counters and progress
                    progress = current_iteration / total_iterations
                    update_display(year, month, state, commodity, progress)

                    print('hello')
                    # Update current iteration
                    current_iteration += 1

# Ensure Pygame quits and program exits safely
pygame.quit()
sys.exit()
```

```python
import os
import pygame
import pyautogui
import time
import sys
from tqdm import tqdm

time.sleep(10)

# Set the window position to top-right
window_width = 400
window_height = 200
screen_width = 2560
screen_height = 1600

x_pos = screen_width - window_width
y_pos = 0

os.environ['SDL_VIDEO_WINDOW_POS'] = f"{x_pos},{y_pos}"



# Initialize Pygame
pygame.init()

# Set the window size and position
win = pygame.display.set_mode((400, 200))  # Increased size for more progress bars
pygame.display.set_caption("Progress Tracker")

# Define colors
WHITE = (255, 255, 255)
BLACK = (0, 0, 0)
GREEN = (0, 255, 0)

# Set up font
font = pygame.font.SysFont("Arial", 18)

# Function to update display with multiple progress bars
def update_display(year, month, state, commodity, overall_progress):
    win.fill(WHITE)
    
    # Year progress bar
    year_progress = year / 24
    pygame.draw.rect(win, BLACK, (10, 10, 380, 20), 2)  # Outline
    pygame.draw.rect(win, GREEN, (10, 10, 380 * year_progress, 20))  # Fill
    year_text = font.render(f"Year: {year}/24", True, BLACK)
    win.blit(year_text, (10, 35))

    # Month progress bar
    month_progress = month / 12
    pygame.draw.rect(win, BLACK, (10, 60, 380, 20), 2)  # Outline
    pygame.draw.rect(win, GREEN, (10, 60, 380 * month_progress, 20))  # Fill
    month_text = font.render(f"Month: {month}/12", True, BLACK)
    win.blit(month_text, (10, 85))

    # State progress bar
    state_progress = state / 26
    pygame.draw.rect(win, BLACK, (10, 110, 380, 20), 2)  # Outline
    pygame.draw.rect(win, GREEN, (10, 110, 380 * state_progress, 20))  # Fill
    state_text = font.render(f"State: {state}/26", True, BLACK)
    win.blit(state_text, (10, 135))

    # Commodity progress bar
    commodity_progress = commodity / 459
    pygame.draw.rect(win, BLACK, (10, 160, 380, 20), 2)  # Outline
    pygame.draw.rect(win, GREEN, (10, 160, 380 * commodity_progress, 20))  # Fill
    commodity_text = font.render(f"Commodity: {commodity}/459", True, BLACK)
    win.blit(commodity_text, (10, 185))

    # Overall progress bar
    pygame.draw.rect(win, BLACK, (10, 210, 380, 20), 2)  # Outline
    pygame.draw.rect(win, GREEN, (10, 210, 380 * overall_progress, 20))  # Fill
    overall_text = font.render(f"Overall Progress: {overall_progress*100:.2f}%", True, BLACK)
    win.blit(overall_text, (10, 235))

    pygame.display.flip()

# Delay to prepare for the script to start
time.sleep(5)

# Calculate total number of iterations for progress tracking
total_iterations = 24 * 12 * 26 * 459
current_iteration = 0

# Main loop to handle events
running = True
while running:
    for event in pygame.event.get():
        if event.type == pygame.QUIT:
            running = False

    # Nested loops with tqdm progress tracking
    for year in tqdm(range(1, 25), desc="Year Progress"):
        for month in range(1, 13):
            for state in range(1, 27):
                for commodity in range(1, 460):
                    # Calculate overall progress
                    overall_progress = current_iteration / total_iterations

                    # Update Pygame display with current loop counters and progress
                    update_display(year, month, state, commodity, overall_progress)

                    # Your existing pyautogui operations go here
                    # Year Block
                    pyautogui.scroll(5)
                    time.sleep(0.1)
                    pyautogui.click(808, 645)
                    time.sleep(0.1)
                    for _ in range(year):
                        pyautogui.press('down')
                        time.sleep(0.5)
                    pyautogui.press('enter')
                    time.sleep(1)

                    # Month Block
                    pyautogui.scroll(5)
                    time.sleep(0.1)
                    pyautogui.click(809, 695)
                    time.sleep(0.1)
                    for _ in range(month):
                        pyautogui.press('down')
                        time.sleep(0.01)
                    pyautogui.press('enter')
                    time.sleep(5)

                    # State Block
                    pyautogui.scroll(5)
                    time.sleep(0.1)
                    pyautogui.click(792, 744)
                    time.sleep(0.1)
                    for _ in range(state):
                        pyautogui.press('down')
                        time.sleep(0.01)
                    pyautogui.press('enter')
                    time.sleep(3)

                    # Commodity Block
                    pyautogui.scroll(5)
                    time.sleep(0.1)
                    pyautogui.click(774, 790)
                    time.sleep(0.1)
                    for _ in range(commodity):
                        pyautogui.press('down')
                        time.sleep(0.01)
                    pyautogui.press('enter')
                    time.sleep(3)

                    # Submit Block
                    pyautogui.scroll(5)
                    time.sleep(0.1)
                    pyautogui.click(735, 844)
                    time.sleep(4)

                    # Download and Back Block
                    pyautogui.scroll(5)
                    time.sleep(0.1)
                    pyautogui.click(987, 592)
                    time.sleep(5)
                    pyautogui.click(1114, 588)
                    time.sleep(5)

                    # Update current iteration
                    current_iteration += 1

# Ensure Pygame quits and program exits safely
pygame.quit()
sys.exit()

```
